```python
def exp_work(arg):
    years = 0
    months = 0
    month_list= ['', '', '']
    year_list = ['', '', '']
    if arg is np.nan or arg == ' ':
        return None
    arg_split = arg.split()[:7]
    for index, item  in enumerate(arg_split):
        if item in year_list:
            years = int(arg_split[index-1])
        if item in month_list:
            months = int(arg_split[index-1])
    return int(years*12 + months)

print('    :', df[' '].isnull().sum())
df['  ()'] = df[' '].apply(exp_work)
print('   ( ):', df['  ()'].median())
df = df.drop(' ', axis=1)
print('     ():', df['  ()'].isnull().sum())
```

```python
million_cities = ['', '',' ','',
                  '','', '', '--', '',
                  '', '', '','']
def get_city(s):
  l = s.split(',')
  if l[0].strip() == '':
    return ''
  elif l[0].strip() == '-':
    return '-'
  elif l[0].strip() in million_cities:
    return '-'
  else:
    return ''

def get_ready_to_move(arg):
    if ('   ' in arg) or ('   ' in arg):
        return False
    elif '' in arg:
        return True
    else:
        return True

def get_ready_for_bisiness_trips(arg):
    if ('' in arg):
        if ('   ' in arg) or('   ' in arg):
            return False
        else:
            return True
    else:
        return False

df[''] = df[', , '].apply(get_city)
print(' ,   -:',
      round(len(df[df['']=='-']) / len(df) * 100))
df['  '] = df[', , '].apply(get_ready_to_move)
df['  '] = df[', , '].apply(get_ready_for_bisiness_trips)
print(' ,     ,   :',
      round(df[df['  '] & df['  ']].shape[0] / df.shape[0] * 100))
df = df.drop(', , ', axis=1)
```

```python
def get_salary_num(arg):
    salary = float(arg.split()[0])
    return salary

def get_salary_urrency(arg):
    currency_dict = {
        'USD': 'USD', 'KZT': 'KZT',
        '': 'UAH', '': 'BYN',
        'EUR': 'EUR', 'KGS': 'KGS',
        '': 'UZS', 'AZN': 'AZN'
    }
    curr = arg.split()[1].replace('.', '')
    if curr == '':
        return 'RUB'
    else:
        return currency_dict[curr]

ER = pd.read_csv('data/ExchangeRates.zip')
ER['date'] = pd.to_datetime(ER['date'], format='mixed', dayfirst=True).dt.date
df[' '] = pd.to_datetime(df[' '], format='%d.%m.%Y %H:%M').dt.date

df['_digit'] = df[''].apply(get_salary_num)
df['name_currency'] = df[''].apply(get_salary_urrency)
merged = df.merge(
    ER,
    left_on=['name_currency', ' '],
    right_on=['currency', 'date',],
    how='left'
)
merged['close'] = merged['close'].fillna(1)
merged['proportion'] = merged['proportion'].fillna(1)
df[' ()'] = merged['close'] * merged['_digit'] / merged['proportion']
df = df.drop(['', '_digit', 'name_currency'], axis=1)
print('    ', round(df[' ()'].median()/1000), '. .')
```

```python
def outliers_z_score_mod(data, feature, left=3.0, right=3.0, log_scale=False):
    if log_scale:
        x = np.log(data[feature]+1)
    else:
        x = data[feature]
    mu = x.mean()
    sigma = x.std()
    lower_bound = mu - left * sigma
    upper_bound = mu + right * sigma
    outliers = data[(x < lower_bound) | (x > upper_bound)]
    cleaned = data[(x >= lower_bound) & (x <= upper_bound)]
    return outliers, cleaned

df_clean['_log'] = df_clean[''].apply(lambda x: np.log(x) + 1)

mu = df_clean['_log'].mean()
sigma = df_clean['_log'].std()
lower_bound = mu - 3 * sigma
upper_bound = mu + 4 * sigma

fig = px.histogram(
    df_clean,
    x='_log',
    title='     ',
    width=1200
                  )

fig.add_vline(x=mu,
              line_dash='solid',
              line_color='red'
              )
fig.add_vline(x=lower_bound,
              line_dash='dash',
              line_color='black'
              )
fig.add_vline(x=upper_bound,
              line_dash='dash',
              line_color='black'
              )

fig.show('png')
```
